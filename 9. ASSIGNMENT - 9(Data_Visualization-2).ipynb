```python
titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
titanic.head()
```

### 1. Create a pie chart presenting the male/female proportion

```python
gender_count = titanic['sex'].value_counts()
gender_count
```

```python
gender_count.plot.pie(labels = titanic['sex'].unique(), autopct = '%1.1f%%')
plt.show()
```

### Alternate

```python
# 1st
gender = titanic['sex'].astype('category')
gender = gender.cat.codes
gender.head()

plt.pie(gender.value_counts(), labels = titanic['sex'].unique(), autopct = '%1.1f%%')
plt.show()
```

### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.figure(figsize = (8, 5))
plt.scatter(titanic['age'], titanic['fare'], c = gender)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare-age graph(gender bais)')
```
